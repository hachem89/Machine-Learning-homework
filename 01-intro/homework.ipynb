```python
# Step 1: Filter Innjoo laptops
df_innjoo = df[df["Brand"] == "Innjoo"]

# Step 2: Select columns
X = df_innjoo[["RAM", "Storage", "Screen"]].values
print(X)
```

```python
# Step 3: Compute XTX
XTX = X.T @ X

# Step 4: Inverse of XTX
XTX_inv = np.linalg.inv(XTX)

# Step 5: Create y
y = np.array([1100, 1300, 800, 900, 1000, 1100])

# Step 6: Compute w
w = XTX_inv @ X.T @ y

# Step 7: Sum of elements
print(w.sum())
```
